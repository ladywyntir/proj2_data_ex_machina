```python
# dependencies
import requests
from config import api_key
import time
import pandas as pd
import warnings
warnings.filterwarnings(action='once')
```

```python
# this is the beginning of the URL to retrieve the genre data
genre_url = "https://api.themoviedb.org/3/genre/movie/list?api_key="

# build the URL
themoviedb_query = f"{genre_url}{api_key}" 
print(themoviedb_query)
```

```python
# attempt to retrieve the data
movies = requests.get(themoviedb_query).json()
movies
```

```python
# now we know the genre we're looking for is named "Science Fiction" and it has an ID of 878
# we'll try to pull the movies with this genre

# base URL for the general movie DB
base_url = "https://api.themoviedb.org/3/discover/movie?api_key="
# "https://api.themoviedb.org/3/movie/550?api_key="

# search for the specific genre
genre_search = "878"

# search for language (international brought over 700 pages of results)
# original_language = "en"
# &with_original_language={original_language}

# search with a beginning and end date
# start_date = "20020101"
# end_date = "20220906"


# build the URL NOTE: this was commented out because I used a for loop to go through all 775 pages, see below
scifi_query = f"{base_url}{api_key}&with_genres={genre_search}"
print(scifi_query)
```

```python
# let's see if we can retrieve some data!
genre_output = requests.get(scifi_query).json()
genre_output

# total page count: 775
# total record count: +15,000
```

```python
# gotta get all the pages throw 'total_pages': 775 in a for loop?
test_movie_DF = pd.DataFrame(columns=["title", "vote_average", "vote_count", "original_language"])

for page in range(1, 775):
    scifi_query = f"{base_url}{api_key}&with_genres={genre_search}&page={page}"
    genre_output = requests.get(scifi_query).json()
    for scifi_movies in genre_output['results']:

        movie_title = scifi_movies['title']
        vote_avg = scifi_movies["vote_average"]
        vote_count = scifi_movies["vote_count"]
        orig_lang = scifi_movies['original_language']     

        # now we throw these results into a DataFrame so we can manipulate it!

        test_movie_DF = test_movie_DF.append({"title": movie_title, "vote_average": vote_avg, "vote_count": vote_count,
                                              "original_language": orig_lang}, ignore_index=True)
```
